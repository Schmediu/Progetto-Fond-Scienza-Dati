```python
import numpy as np
import pandas as pd
```

```python
#dataset sui film
df_uncut = pd.read_csv("dataset/TMDB_movie_dataset_v11.csv", encoding = "utf8")
df_0 = df_uncut.drop(columns = ["backdrop_path", "id", "vote_average", "original_language", "production_companies", "poster_path", "tagline", "spoken_languages", "imdb_id", "homepage", "overview", "adult", "status", "popularity", "keywords", "original_title"])
df_1 = df_0[~(df_0["budget"] == 0) & ~(df_0["revenue"] == 0) & ~(df_0["budget"].isnull()) & ~(df_0["revenue"].isnull()) & ~(df_0["release_date"].isnull()) & ~(df_0["genres"].isnull()) & ~(df_0["production_countries"].isnull())]
df = df_1.reset_index(drop=True)
df
```

```python
#dataset sull'inflazione
inflation = pd.read_csv("dataset/inflation.csv", encoding = "utf8")
inflation.set_index("Unnamed: 0", inplace = True)
inflation["mean"] = inflation[["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]].mean(axis = 1)
inflation_final = inflation.drop(columns= ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
inflation_final
```

```python
#trasformo l'anno da float a int
for i in range(110):
    inflation_final.loc[i, "Year"] == int(inflation_final.loc[i, "Year"])
inflation_final
```

```python
#avere la data di uscita espessa solo con l'anno
for i in range(11027):
    data = str(df.loc[i, "release_date"])[:4]
    df.loc[i, "release_date"] = int(data)
df
    
```

```python
#Modifico valori errati del dataset
df.loc[4549, "budget"] = 1034000
df.loc[8604, "budget"] = 40000
df.loc[5814, "budget"] = 1500000
df.loc[9080, "budget"] = 903557
df = df.reset_index(drop = True)
df
```

```python
#unione dei due dataset così da avere il valore "inflation" per ogni anno
for i in range(11027):
    for a in range(110):
        if str(df.loc[i, "release_date"]) == str(inflation_final.loc[a, "Year"])[:4]:
            df.loc[i, "inflation"] = int(inflation_final.loc[a, "mean"])
df
            
```

```python
#calcolare budget e revenue tenendo conto dell'inflazione
df["inflated_budget"] = df["budget"] * 100 // df["inflation"]
df["inflated_revenue"] = df["revenue"] * 100 // df["inflation"]
df
```

```python
#vedere tutti i film di animazione:
df_animation = df
for i in range(11027):
    lista = df["genres"][i].split(",")
    lista = [f.strip() for f in lista]
    if "Animation" not in lista:
        df_animation = df_animation.drop(labels = i, axis = 0)
df_animation
```

```python
#tolgo i film successivi al 2022 in quanto privi di dato sull'inflazione
df_final = df[ ~(df["release_date"] > 2022)]
df_final
```

```python
#resetto l'indice
df_0 = df_final.reset_index(drop=True)
df_0
```

```python
#aggiungo la colonna della percentuale del ricavato in base al budget
for i in range(10663):
    df_0.loc[i, "percentual_of_revenue"] = df_0.loc[i, "revenue"] / df_0.loc[i, "budget"] * 100
df_0
```

```python
#modifico errori riscontrati nel dataframe 
df_1 = df_0.drop(labels = [9335, 9328, 9558, 10147, 7116, 7606, 9807, 9973], axis = 0)
df_2 = df_1.reset_index(drop = True)
df_2
```
